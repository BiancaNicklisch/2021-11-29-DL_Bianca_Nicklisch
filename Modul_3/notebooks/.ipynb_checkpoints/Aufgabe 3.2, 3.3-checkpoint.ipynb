```python
import pandas as pd
df = pd.read_csv("../data/Library_Usage.csv")
df['Total Checkouts'].mean()
```

```python
df['Total Checkouts'].median()
```

```python
    #Schau Dir den Mittelwert und den Median der Variable Total Checkouts an. Warum sind die beiden Werte so unterschiedlich?
    Der Mittelwert gibt den Durschnitt der Ausleihen an alle Personen an.
    Der Median gibt den Wert zurück, der genau in der Mitte der numerisch sortierten Ausleihen steht.
    
    #Was ziehst Du daraus für Schlüsse für weitere statistische Analysen und Reports?
    Der Mittelwert sehr stark beeinflusst von extremen oder hahen werten wo hingegen der Medianwert relativ stabil bleibt.
```

```python
df['Total Checkouts'].quantile(q=[0.25, 0.5, 0.75])
```

```python
alpha = 0.005
df['Total Checkouts'].quantile([alpha, 1-alpha])
```

```python
alpha = 0.015
quantile = df['Total Checkouts'].quantile([alpha, 1-alpha])
print(quantile)
```

```python
df_gefiltert1 = df[(df['Total Checkouts'] <= quantile[alpha])]
df_gefiltert1 
```

```python
df_gefiltert2 = df[(df['Total Checkouts'] >= quantile[1-alpha])]
df_gefiltert2
```

```python
df_gefiltert3 = df[(df['Total Checkouts'] > quantile[alpha]) & (df['Total Checkouts'] < quantile[1-alpha])]
df_gefiltert3 
```
